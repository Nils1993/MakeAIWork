```python
import requests
import pandas as pd
import json
import time
```

```python
# get data from URL
url = "http://localhost:8080/medish_centrum_randstad/api/netlify?page=1"

data = []

while True:
    r = requests.get(url)
    if r.status_code != 200:
        continue

    if r.headers["Content-Type"] == 'application/json':
        data_json = json.loads(r.text)
        print(len(data_json))
        data.extend(data_json['data'])
    if 'next' in data_json:
        url = data_json['next']
        time.sleep(0.3)
    else:
        break
```

```python
# Convert json data to pandas
df = pd.DataFrame(data)
display(df)
```

```python
# Save dataframe as csv.

df.to_csv('C:/Users/nilsm/MakeAIWork/projects/project1/csv_files\data_frame.csv')
```
