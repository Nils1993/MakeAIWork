```python
import matplotlib.pyplot as plt
import pandas as pd

file = pd.read_csv('C:/Users/nilsm/MakeAIWork/data/csv/bijdrage-van-het-toerisme-aan-de-totale-werkgelegenheid__contribution-of-tourism-to-total-employment.csv')



file.head(30)
```

```python
d_file = file[['NAME_HTML', 'VALUE']].copy()

d_file.head(30)
```

```python
d_file.sort_values(by = ['VALUE'], inplace = True)

d_file_oh = pd.get_dummies(d_file)

d_file_oh.head(30)

# d_file_oh.plot('NAME_HTML', 'VALUE', label='work by tourism', kind='bar')
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt


clusters = shc.linkage(d_file_oh, 
            method='ward', 
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering_model = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
clustering_model.fit(d_file_oh)
clustering_model.labels_
```

```python
import seaborn as sns

sns.barplot(x='NAME_HTML', 
                y='VALUE', 
                data=d_file, 
                hue=clustering_model.labels_,
                palette="rainbow").set_title('Labeled Customer Data')
```
