## Opdracht 1: inlezen

Lees de Comodity Prices dataset in.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read csv file

# ik gebruik de .read_csv() functie
df = pd.read_csv("./data/commodity_prices.csv")
```

## Opdracht 2: exploreren

Welke kolommen bevat deze dataset?

```python
# Print info

df
```

Welk datatype heeft elke kolom?

```python
# Beschrijf datatype
# ik gebruik .dtypes om de datatype van iedere column te weergeven
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dtypes.html
df.dtypes
```

Wat is de gemiddelde prijs voor ruwe olie over de jaren?

```python
# Zoek in de Pandas documentatie een geschikte functie
# Hint: https://pandas.pydata.org/docs/reference/series.html

# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.mean.html
df_crude_oil = df["Crude Oil"]
display(df_crude_oil)

# Voer deze functie uit op het veld 'Crude Oil' en print het resultaat

df_crude_oil.mean()
```

Wanneer was de olieprijs op zijn hoogst? En op zijn laagst?

```python
# Print highest and lowest price

# .max() en .min():
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.max.html
display(df["Crude Oil"].max())

# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.min.html
display(df["Crude Oil"].min())

# Print het jaar van de hoogste en laagste prijs

# om de juiste index te vinden van het minimale en maximale nummer gebruik ik .idxmin() en .idxmax():
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmin.html
display(df.loc[df["Crude Oil"].idxmin(), 'Year'])

# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html
display(df.loc[df["Crude Oil"].idxmax(), 'Year'])
```

Net als SQL kan Pandas ook queries uitvoeren en filteren.

Zoek alle jaren waarin de olieprijs tussen 50 en 80 dollar was.

```python
# Hint: filter jaren waar de prijs tussen 50 en 80 was

# roep op met df[df["column"]], gebruik vervolgens de .between() function.
# https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html
display(
    df[df["Crude Oil"].between(50, 80)]
)

print("\n")

display(
    df["Year"]
    [df["Crude Oil"].between(50, 80)]
)
```

## Opdracht 3: samenstellen

Stel dat Nutella zou bestaan uit 50% cacao, 20% suiker, 5% mais en 25% kokosolie.

Verder is gegeven:

- voor deze grondstoffen is de prijs per kilogram
- een pot bevat 800 gram Nutella

Wat is dan de prijs voor een pot Nutella in elk jaar?

```python
# Voeg een kolom 'Nutella' toe aan het DataFrame:

# gebruik .insert()
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.insert.html
df.insert(1, "Nutella", float, None)

# Bereken de prijs voor een pot van 800 gram over de jaren heen:

# geef een waarde aan 'Nutella' door alle ingredienten bij elkaar op te tellen en vervolgens * 0.8 (kg) te doen
# gebruik round(waarde, 2) om af te ronden tot 2 decimale na de comma.
df['Nutella'] = round(((df['Cocoa'] * 0.5) + (df["Sugar"] * 0.2) + (df["Maize"] * 0.05) + (df["Coconut Oil"] * 0.25)) * 0.8,2)

display(df)
```

Zoals je ziet, ontbreekt de data voor kolen.

Maak op basis van de aanwezige data een schatting van de ontbrekende waarden:

2. Loopt de prijs van kolen op met een andere grondstof?
